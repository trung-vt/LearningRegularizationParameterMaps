```python
import os

def remove_models(folder, condition):
    for root, dirs, files in os.walk(folder):
        for file in files:
            if file.endswith('.pt'):
                # Get the characters after the LAST underscore
                epoch = file.split('_')[-1].split('.')[0]
                # Check if the epoch is a number
                if epoch.isdigit():
                    epoch = int(epoch)
                else:
                    continue
                if condition(epoch):
                    print(f"Removed {file}")
                    os.remove(os.path.join(root, file))



left = 100
right = 5000
step = 100

def to_remove(epoch):
    return epoch > left and epoch < right and epoch % step != 0

# models_path = '../dyn_img_static/tmp_2/'
# model_name = 'model_turtle_2024_06_04_04_19_21'
# models_folder = os.path.join(models_path, model_name)

# models_folder = "../../src/tmp_2/model-chest_xray-2024_06_08_15_37_55"
remove_models(models_folder, to_remove)
```
