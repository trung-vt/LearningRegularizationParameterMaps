```python
import os
import matplotlib.pyplot as plt
from PIL import Image

data_path = "images/patches"
files = os.listdir(data_path)

examples = [
    "0064_GT_SRGB_010_10",
    "0064_GT_SRGB_010_23",
    "0065_GT_SRGB_011_16",
    "0075_GT_SRGB_010_20",
    "0098_GT_SRGB_010_19",
    "0156_GT_SRGB_010_5",
]

plt.figure(figsize=(10, 6))
for i, ex in enumerate(examples):
    example_image = Image.open(f"{data_path}/{ex}.png")
    plt.subplot(2, 3, i+1)
    plt.imshow(example_image)
    plt.axis("off")
plt.show();
```
